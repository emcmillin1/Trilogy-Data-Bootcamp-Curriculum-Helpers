## Data Generation

```python
from sklearn.datasets import make_classification
```

```python
data = make_classification(n_samples=1000,
                           n_features=5,
                           n_informative=4,
                           n_redundant=1,
                           n_classes=2,
                           random_state=42)[0]

df = pd.DataFrame(
        data, 
        columns=['a','b','c','d','e'])
    
df.head()
```

## Plotting

```python
# option 1: plt.plot
plt.scatter(df['a'], df['b'], color='orange', marker='x', s=(df['c'])*2)
plt.title("Some title")
plt.xlabel("X axis")
plt.ylabel("y axis")
```
